# Data Scraping via Spotify Web API

```python
import pandas as pd
import spotipy
import pprint
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
cid ="32093566a32d4f7b8cdd6531435cfc29"  #Client ID (PLease use your own spotify dev account to request)
secret = "e2a2881fbdc848aabb1558489c136c39"

#Initialize Spotify Client
client_credentials_manager = SpotifyClientCredentials(client_id=cid, client_secret=secret)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

```python
playlist_link = "https://open.spotify.com/playlist/6UeSakyzhiEt4NB3UAd6NQ?si=f4d76ac5c03848cb" #Billboard top 100 songs playlist (This playlist is updated weekly)

#Extract the playlist URI from link
playlist_URI = playlist_link.split("/")[-1].split("?")[0]

#Retrieve the URIs of all the tracks
track_uris = [x["track"]["uri"] for x in sp.playlist_tracks(playlist_URI)["items"]]
```

```python
pp = pprint.PrettyPrinter()

#Displaying the tracks to check & compare the playlist
pp.pprint(track_uris) #Top 100 tracks
```

```python
#Create empty lists to store audio features
song_names = []
song_popularity = []
song_duration_ms = []
acousticness = []
danceability = []
energy = []
instrumentalness = []
key = []
liveness = []
loudness = []
audio_mode = []
speechiness = []
tempo = []
time_signature = []
audio_valence = []

#Loop through the list of track URIs and retrieve audio features
for track_uri in track_uris:
    track_audio_features = sp.audio_features(track_uri)[0]

    #Extract audio features
    song_names.append(sp.track(track_uri)["name"])
    song_popularity.append(sp.track(track_uri)["popularity"])
    song_duration_ms.append(track_audio_features["duration_ms"])
    acousticness.append(track_audio_features["acousticness"])
    danceability.append(track_audio_features["danceability"])
    energy.append(track_audio_features["energy"])
    instrumentalness.append(track_audio_features["instrumentalness"])
    key.append(track_audio_features["key"])
    liveness.append(track_audio_features["liveness"])
    loudness.append(track_audio_features["loudness"])
    audio_mode.append(track_audio_features["mode"])
    speechiness.append(track_audio_features["speechiness"])
    tempo.append(track_audio_features["tempo"])
    time_signature.append(track_audio_features["time_signature"])
    audio_valence.append(track_audio_features["valence"])

#Create dataframe to store it
features = {
    'song_name': song_names,
    'song_popularity': song_popularity,
    'song_duration_ms': song_duration_ms,
    'acousticness': acousticness,
    'danceability': danceability,
    'energy': energy,
    'instrumentalness': instrumentalness,
    'key': key,
    'liveness': liveness,
    'loudness': loudness,
    'audio_mode': audio_mode,
    'speechiness': speechiness,
    'tempo': tempo,
    'time_signature': time_signature,
    'audio_valence': audio_valence
}

df = pd.DataFrame(features)
```
